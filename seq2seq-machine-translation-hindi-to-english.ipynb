```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input,SimpleRNN, Embedding, LSTM, Dense, Bidirectional, 
    Attention, Concatenate, Dropout, Add, LayerNormalization,
    MultiHeadAttention, TimeDistributed
)
from tensorflow.keras.callbacks import EarlyStopping, Callback
from sklearn.model_selection import train_test_split
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge_score import rouge_scorer
import matplotlib.pyplot as plt
from tqdm import tqdm
import gc

# Memory management
tf.config.optimizer.set_jit(True)
tf.keras.backend.clear_session()


# ====================================================================
# 1. IMPORT AND SETUP
# ====================================================================
import gc

# Memory management
tf.config.optimizer.set_jit(True)
tf.keras.backend.clear_session()
gc.collect()

# ====================================================================
# 2. LOAD DATA
# ====================================================================
def load_data(filepath, max_samples=100000):
    """Load Hindi-English parallel corpus from CSV file
    
    Args:
        filepath: Path to CSV file with 'hindi' and 'english' columns
        max_samples: Maximum number of samples to use
    
    Returns:
        hindi_sentences, english_sentences as lists
    """
    print(f"📂 Loading data from: {filepath}")
    
    # Try multiple encodings
    encodings = ['utf-8', 'latin-1', 'iso-8859-1', 'cp1252']
    df = None
    
    for encoding in encodings:
        try:
            df = pd.read_csv(filepath, encoding=encoding)
            print(f"✅ Successfully loaded with encoding: {encoding}")
            break
        except (UnicodeDecodeError, UnicodeError):
            continue
    
    if df is None:
        raise ValueError(f"Could not read {filepath} with any encoding. Tried: {encodings}")
    
    print(f"📊 Total rows: {len(df)}")
    print(f"📋 Columns: {df.columns.tolist()}")
    
    # Check for required columns
    if 'hindi' not in df.columns or 'english' not in df.columns:
        raise ValueError(f"CSV must have 'hindi' and 'english' columns. Found: {df.columns.tolist()}")
    
    # Drop missing values
    df = df.dropna(subset=['hindi', 'english'])
    print(f"✅ Valid rows (after removing NaN): {len(df)}")
    
    # Sample if needed
    if len(df) > max_samples:
        df = df.sample(n=max_samples, random_state=42)
        print(f"✅ Sampled {max_samples} rows for training")
    
    # Convert to lists
    hindi_sentences = df['hindi'].astype(str).tolist()
    english_sentences = df['english'].astype(str).tolist()
    
    # Show samples
    print(f"\n📋 Sample data:")
    for i in range(min(3, len(hindi_sentences))):
        hi = hindi_sentences[i][:80] + '...' if len(hindi_sentences[i]) > 80 else hindi_sentences[i]
        en = english_sentences[i][:80] + '...' if len(english_sentences[i]) > 80 else english_sentences[i]
        print(f"  [{i+1}] HI: {hi}")
        print(f"      EN: {en}")
    
    return hindi_sentences, english_sentences


# ====================================================================
# LOAD YOUR DATASET - CHANGE THE PATH HERE
# ====================================================================
# For Kaggle, use the path to your uploaded dataset
# Example: '/kaggle/input/your-dataset-name/hindi_english_parallel.csv'

DATASET_PATH = '/kaggle/input/hindi-english-parallel-corpus/hindi_english_parallel.csv'
# Or if file is in current directory: 'hindi_english_parallel.csv'

hindi_data, english_data = load_data(DATASET_PATH, max_samples=8000)
print(f"\n✅ Dataset loaded: {len(hindi_data)} sentence pairs")


# ====================================================================
# 3. PREPROCESSING
# ====================================================================
def preprocess_text(sentences):
    """Add start and end tokens"""
    processed = []
    for s in sentences:
        text = str(s).lower().strip()
        # Remove extra spaces
        text = ' '.join(text.split())
        processed.append("starttoken " + text + " endtoken")
    return processed


hindi_processed = preprocess_text(hindi_data)
english_processed = preprocess_text(english_data)

# FIXED: Remove apostrophe from filters (it's important for contractions)
hin_tokenizer = Tokenizer(filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', oov_token='<OOV>')
hin_tokenizer.fit_on_texts(hindi_processed)
eng_tokenizer = Tokenizer(filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', oov_token='<OOV>')
eng_tokenizer.fit_on_texts(english_processed)

# Convert to sequences
hin_seq = hin_tokenizer.texts_to_sequences(hindi_processed)
eng_seq = eng_tokenizer.texts_to_sequences(english_processed)

# IMPROVED: Use percentile-based max length to handle outliers
max_hin_len = int(np.percentile([len(x) for x in hin_seq], 95))
max_eng_len = int(np.percentile([len(x) for x in eng_seq], 95))
hin_vocab = len(hin_tokenizer.word_index) + 1
eng_vocab = len(eng_tokenizer.word_index) + 1

print(f"📊 Hindi vocab size: {hin_vocab}")
print(f"📊 English vocab size: {eng_vocab}")
print(f"📊 Max Hindi length: {max_hin_len}")
print(f"📊 Max English length: {max_eng_len}")

# Pad sequences
hin_pad = pad_sequences(hin_seq, maxlen=max_hin_len, padding='post', truncating='post')
eng_pad = pad_sequences(eng_seq, maxlen=max_eng_len, padding='post', truncating='post')

# CRITICAL FIX: Proper decoder input/output creation
dec_in = eng_pad[:, :-1]  # All tokens except last
dec_out = eng_pad[:, 1:]  # All tokens except first

print(f"📊 Encoder input shape: {hin_pad.shape}")
print(f"📊 Decoder input shape: {dec_in.shape}")
print(f"📊 Decoder output shape: {dec_out.shape}")

# Train-test split
X_train, X_test, y_train_in, y_test_in, y_train_out, y_test_out = train_test_split(
    hin_pad, dec_in, dec_out, test_size=0.15, random_state=42
)

print(f"✅ Training samples: {len(X_train)}")
print(f"✅ Test samples: {len(X_test)}")


# ====================================================================
# 2. OPTIMIZED CONFIG FOR 25GB RAM
# ====================================================================
EMB_DIM = 256  # Increased from 128
LAT_DIM = 512  # Increased from 256
DROPOUT = 0.3
EPOCHS = 15  # Balanced epochs
BATCH_SIZE = 128  # Larger batch for faster training


# ====================================================================
# 3. OPTIMIZED MODEL DEFINITIONS
# ===================================================================


def build_birnn():
    """Bidirectional RNN Encoder-Decoder"""
    enc_in = Input(shape=(max_hin_len,))
    x = Embedding(hin_vocab, EMB_DIM)(enc_in)
    x = Dropout(DROPOUT)(x)
    enc_out, fh, bh = Bidirectional(
        SimpleRNN(LAT_DIM//2, return_sequences=False, return_state=True, dropout=0.2),
        merge_mode='concat'
    )(x)
    
    # Concatenate forward and backward states
    state_h = Concatenate()([fh, bh])
    
    dec_in = Input(shape=(max_eng_len - 1,))
    y = Embedding(eng_vocab, EMB_DIM)(dec_in)
    y = Dropout(DROPOUT)(y)
    dec_out = SimpleRNN(LAT_DIM, return_sequences=True, dropout=0.2)(y, initial_state=state_h)
    dec_out = Dropout(DROPOUT)(dec_out)
    out = Dense(eng_vocab, activation='softmax')(dec_out)
    return Model([enc_in, dec_in], out)



def build_bilstm():
    """Bidirectional LSTM with 25GB optimization"""
    enc_in = Input(shape=(max_hin_len,))
    x = Embedding(hin_vocab, EMB_DIM)(enc_in)
    x = Dropout(DROPOUT)(x)
    enc_out, fh, fc, bh, bc = Bidirectional(
        LSTM(LAT_DIM//2, return_sequences=False, return_state=True, recurrent_dropout=0.2)
    )(x)
    
    state_h = Concatenate()([fh, bh])
    state_c = Concatenate()([fc, bc])
    
    dec_in = Input(shape=(max_eng_len - 1,))
    y = Embedding(eng_vocab, EMB_DIM)(dec_in)
    y = Dropout(DROPOUT)(y)
    dec_out = LSTM(LAT_DIM, return_sequences=True, recurrent_dropout=0.2)(y, initial_state=[state_h, state_c])
    dec_out = Dropout(DROPOUT)(dec_out)
    out = Dense(eng_vocab, activation='softmax')(dec_out)
    return Model([enc_in, dec_in], out)


def build_bilstm_attention():
    """BiLSTM with Attention - 25GB optimized"""
    enc_in = Input(shape=(max_hin_len,))
    enc_emb = Embedding(hin_vocab, EMB_DIM)(enc_in)
    enc_emb = Dropout(DROPOUT)(enc_emb)
    enc_out, fh, fc, bh, bc = Bidirectional(
        LSTM(LAT_DIM//2, return_sequences=True, return_state=True, recurrent_dropout=0.2)
    )(enc_emb)
    
    state_h = Concatenate()([fh, bh])
    state_c = Concatenate()([fc, bc])
    
    dec_in = Input(shape=(max_eng_len - 1,))
    dec_emb = Embedding(eng_vocab, EMB_DIM)(dec_in)
    dec_emb = Dropout(DROPOUT)(dec_emb)
    dec_out, _, _ = LSTM(LAT_DIM, return_sequences=True, return_state=True, recurrent_dropout=0.2)(
        dec_emb, initial_state=[state_h, state_c]
    )
    
    # Attention mechanism
    attn_out = Attention()([dec_out, enc_out])
    comb = Concatenate(axis=-1)([dec_out, attn_out])
    comb = Dropout(DROPOUT)(comb)
    out = Dense(eng_vocab, activation='softmax')(comb)
    return Model([enc_in, dec_in], out)


def positional_encoding(length, depth):
    """Generate positional encodings"""
    depth = depth / 2
    positions = np.arange(length)[:, np.newaxis]
    depths = np.arange(depth)[np.newaxis, :] / depth
    angle_rates = 1 / (10000**depths)
    angle_rads = positions * angle_rates
    pos_encoding = np.concatenate([np.sin(angle_rads), np.cos(angle_rads)], axis=-1)
    return tf.cast(pos_encoding, dtype=tf.float32)


def build_transformer():
    """Transformer with 25GB optimization"""
    NUM_HEADS = 4  # Increased from 4
    FFN_DIM = 256  # Increased from 256
    NUM_LAYERS = 2  # Increased from 2
    
    # Encoder
    enc_in = Input(shape=(max_hin_len,), name='encoder_input')
    enc_emb = Embedding(hin_vocab, EMB_DIM)(enc_in)
    pos_enc = positional_encoding(max_hin_len, EMB_DIM)
    enc_emb = enc_emb + pos_enc
    enc_emb = Dropout(DROPOUT)(enc_emb)
    
    enc_output = enc_emb
    for i in range(NUM_LAYERS):
        attn_out = MultiHeadAttention(
            num_heads=NUM_HEADS, 
            key_dim=EMB_DIM // NUM_HEADS,
            dropout=DROPOUT,
            name=f'enc_mha_{i}'
        )(enc_output, enc_output)
        attn_out = Dropout(DROPOUT)(attn_out)
        enc_output = Add()([enc_output, attn_out])
        enc_output = LayerNormalization(epsilon=1e-6)(enc_output)
        
        ffn_out = Dense(FFN_DIM, activation='relu')(enc_output)
        ffn_out = Dropout(DROPOUT)(ffn_out)
        ffn_out = Dense(EMB_DIM)(ffn_out)
        ffn_out = Dropout(DROPOUT)(ffn_out)
        enc_output = Add()([enc_output, ffn_out])
        enc_output = LayerNormalization(epsilon=1e-6)(enc_output)
    
    # Decoder
    dec_in = Input(shape=(max_eng_len - 1,), name='decoder_input')
    dec_emb = Embedding(eng_vocab, EMB_DIM)(dec_in)
    pos_dec = positional_encoding(max_eng_len - 1, EMB_DIM)
    dec_emb = dec_emb + pos_dec
    dec_emb = Dropout(DROPOUT)(dec_emb)
    
    dec_output = dec_emb
    for i in range(NUM_LAYERS):
        self_attn = MultiHeadAttention(
            num_heads=NUM_HEADS, 
            key_dim=EMB_DIM // NUM_HEADS,
            dropout=DROPOUT,
            name=f'dec_self_mha_{i}'
        )(dec_output, dec_output, use_causal_mask=True)
        self_attn = Dropout(DROPOUT)(self_attn)
        dec_output = Add()([dec_output, self_attn])
        dec_output = LayerNormalization(epsilon=1e-6)(dec_output)
        
        cross_attn = MultiHeadAttention(
            num_heads=NUM_HEADS, 
            key_dim=EMB_DIM // NUM_HEADS,
            dropout=DROPOUT,
            name=f'dec_cross_mha_{i}'
        )(dec_output, enc_output)
        cross_attn = Dropout(DROPOUT)(cross_attn)
        dec_output = Add()([dec_output, cross_attn])
        dec_output = LayerNormalization(epsilon=1e-6)(dec_output)
        
        ffn_out = Dense(FFN_DIM, activation='relu')(dec_output)
        ffn_out = Dropout(DROPOUT)(ffn_out)
        ffn_out = Dense(EMB_DIM)(ffn_out)
        ffn_out = Dropout(DROPOUT)(ffn_out)
        dec_output = Add()([dec_output, ffn_out])
        dec_output = LayerNormalization(epsilon=1e-6)(dec_output)
    
    out = Dense(eng_vocab, activation='softmax', name='output')(dec_output)
    return Model([enc_in, dec_in], out)


# ====================================================================
# 4. CRITICAL FIX: PROPER INFERENCE FUNCTION
# ====================================================================
def decode_sequence_greedy(model, input_seq, max_length):
    """FIXED: Proper greedy decoding with autoregressive generation"""
    # Get start token id
    start_token_id = eng_tokenizer.word_index.get('starttoken', 1)
    end_token_id = eng_tokenizer.word_index.get('endtoken', 2)
    
    # Initialize decoder input with start token
    decoder_input = np.zeros((1, max_length))
    decoder_input[0, 0] = start_token_id
    
    output_sentence = []
    
    for i in range(1, max_length):
        # Predict next token
        predictions = model.predict([input_seq, decoder_input], verbose=0)
        predicted_id = np.argmax(predictions[0, i-1, :])
        
        # Stop if end token
        if predicted_id == end_token_id or predicted_id == 0:
            break
        
        # Add predicted token to output
        word = eng_tokenizer.index_word.get(predicted_id, '')
        if word and word not in ['starttoken', 'endtoken']:
            output_sentence.append(word)
        
        # Update decoder input for next iteration
        if i < max_length:
            decoder_input[0, i] = predicted_id
    
    return ' '.join(output_sentence)


# ====================================================================
# 5. COMPILATION AND TRAINING
# ====================================================================
def compile_model(m):
    m.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return m


# Build and train models (optimized for 25GB)
models_to_train = {
    'BiRNN': build_birnn(),
    'BiLSTM': build_bilstm(),
    'BiLSTM+Attention': build_bilstm_attention(),
    'Transformer': build_transformer()  # Now included
}

for name in models_to_train:
    models_to_train[name] = compile_model(models_to_train[name])

# Expand dims for sparse categorical crossentropy
y_train_exp = np.expand_dims(y_train_out, -1)
y_test_exp = np.expand_dims(y_test_out, -1)

# Training with progress tracking
for name, model in models_to_train.items():
    print(f"\n{'='*60}")
    print(f"🚀 Training {name}")
    print(f"{'='*60}")
    
    # Show model summary
    print(f"Model parameters: {model.count_params():,}")
    
    history = model.fit(
        [X_train, y_train_in], 
        y_train_exp,
        validation_data=([X_test, y_test_in], y_test_exp),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        verbose=2,  # Less verbose for faster execution
        callbacks=[
            EarlyStopping(
                patience=7,  # Increased patience
                restore_best_weights=True,
                monitor='val_loss',
                verbose=1
            ),
            tf.keras.callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=4,  # Increased patience
                min_lr=0.00001,
                verbose=1
            )
        ]
    )
    
    print(f"✅ {name} - Final train loss: {history.history['loss'][-1]:.4f}, "
          f"val loss: {history.history['val_loss'][-1]:.4f}, "
          f"val acc: {history.history['val_accuracy'][-1]:.4f}")
    
    # Clear memory after each model
    tf.keras.backend.clear_session()
    gc.collect()


# ====================================================================
# 6. FAST EVALUATION WITH BATCHING
# ====================================================================
def decode_sequence_batch(model, input_seqs, max_length, batch_size=32):
    """Batch decoding for faster evaluation"""
    start_token_id = eng_tokenizer.word_index.get('starttoken', 1)
    end_token_id = eng_tokenizer.word_index.get('endtoken', 2)
    
    results = []
    n_batches = (len(input_seqs) + batch_size - 1) // batch_size
    
    for batch_idx in range(n_batches):
        start_idx = batch_idx * batch_size
        end_idx = min((batch_idx + 1) * batch_size, len(input_seqs))
        batch_input = input_seqs[start_idx:end_idx]
        
        # Initialize decoder inputs
        decoder_inputs = np.zeros((len(batch_input), max_length))
        decoder_inputs[:, 0] = start_token_id
        
        for i in range(1, max_length):
            predictions = model.predict([batch_input, decoder_inputs], verbose=0, batch_size=batch_size)
            predicted_ids = np.argmax(predictions[:, i-1, :], axis=-1)
            
            # Check for end tokens
            if i < max_length:
                decoder_inputs[:, i] = predicted_ids
        
        # Convert to sentences
        for dec_seq in decoder_inputs:
            sentence = []
            for token_id in dec_seq[1:]:  # Skip start token
                if token_id == end_token_id or token_id == 0:
                    break
                word = eng_tokenizer.index_word.get(token_id, '')
                if word and word not in ['starttoken', 'endtoken']:
                    sentence.append(word)
            results.append(' '.join(sentence))
    
    return results


def compute_metrics(model, n_samples=200):
    """Compute BLEU and ROUGE with batch processing"""
    smoothie = SmoothingFunction().method4
    scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
    bleu_scores, rouge_scores = [], []
    
    n = min(n_samples, len(X_test))
    batch_size = 32
    
    print(f"Generating predictions for {n} samples...")
    predictions = decode_sequence_batch(model, X_test[:n], max_eng_len - 1, batch_size=batch_size)
    
    print("Computing metrics...")
    for i in tqdm(range(n), desc="Scoring"):
        # Get reference
        ref_ids = y_test_out[i]
        ref = ' '.join([eng_tokenizer.index_word.get(x, '') 
                       for x in ref_ids if x > 0 and x != eng_tokenizer.word_index.get('endtoken', 2)])
        ref = ref.replace('starttoken', '').strip()
        
        pred = predictions[i]
        
        if pred and ref:
            bleu = sentence_bleu([ref.split()], pred.split(), 
                               smoothing_function=smoothie)
            rouge = scorer.score(ref, pred)['rougeL'].fmeasure
            bleu_scores.append(bleu)
            rouge_scores.append(rouge)
    
    return np.mean(bleu_scores), np.mean(rouge_scores)


# Evaluate models
results = {}
for name, model in models_to_train.items():
    print(f"\n📊 Evaluating {name}...")
    bleu, rouge = compute_metrics(model, n_samples=200)
    results[name] = {'BLEU': bleu, 'ROUGE-L': rouge}
    print(f"   BLEU: {bleu:.4f}, ROUGE-L: {rouge:.4f}")
    
    # Clear memory after evaluation
    gc.collect()


# ====================================================================
# 7. RESULTS VISUALIZATION
# ====================================================================
results_df = pd.DataFrame(results).T
print("\n" + "="*60)
print("FINAL RESULTS")
print("="*60)
print(results_df.round(4))

# Save results
results_df.to_csv('translation_results.csv')

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
results_df.plot(kind='bar', ax=ax, width=0.7, color=['#3498db', '#e74c3c'])
plt.title("Model Comparison: BLEU vs ROUGE-L", fontsize=16, fontweight='bold')
plt.xlabel("Model", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.ylim(0, 1)
plt.legend(loc='upper right', fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('model_comparison.png', dpi=300, bbox_inches='tight')
plt.show()


# ====================================================================
# 8. SAMPLE PREDICTIONS
# ====================================================================
print("\n" + "="*60)
print("SAMPLE TRANSLATIONS")
print("="*60)

# Use the best model (typically BiLSTM+Attention or Transformer)
best_model_name = max(results.items(), key=lambda x: x[1]['BLEU'])[0]
best_model = models_to_train[best_model_name]
print(f"\nUsing best model: {best_model_name}")

for i in range(min(5, len(X_test))):
    # Hindi input
    hin_text = ' '.join([hin_tokenizer.index_word.get(x, '') 
                        for x in X_test[i] if x > 0])
    hin_text = hin_text.replace('starttoken', '').replace('endtoken', '').strip()
    
    # Reference translation
    ref_text = ' '.join([eng_tokenizer.index_word.get(x, '') 
                        for x in y_test_out[i] if x > 0])
    ref_text = ref_text.replace('starttoken', '').replace('endtoken', '').strip()
    
    # Predicted translation
    pred_text = decode_sequence_greedy(best_model, X_test[i:i+1], max_eng_len - 1)
    
    print(f"\n[{i+1}] Hindi:     {hin_text}")
    print(f"    Reference: {ref_text}")
    print(f"    Predicted: {pred_text}")

print("\n" + "="*60)
print("✅ Training and Evaluation Complete!")
print("="*60)
```
